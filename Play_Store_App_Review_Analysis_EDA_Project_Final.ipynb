The Play Store, as one of the largest app marketplaces, is a treasure trove of user-generated reviews for various applications. These reviews not only provide valuable feedback for app developers but also offer insights into user preferences, satisfaction, and pain points. In this exploratory data analysis (EDA) project, we delve into the vast ocean of Play Store app reviews to gain a deeper understanding of the sentiments expressed by users and uncover patterns and trends.

The objective of this project is to conduct a comprehensive analysis of Play Store app reviews, aiming to extract meaningful insights from the textual data. By leveraging EDA techniques, we will explore the dataset's characteristics, perform text preprocessing, and employ various statistical and visual analysis methods to uncover patterns, sentiment distributions, and potential correlations. This analysis will provide app developers, marketers, and decision-makers with actionable information to improve app quality, user experience, and overall satisfaction.

**Key Steps and Methods**:

**Data Acquisition**: We obtain a large dataset of Play Store app reviews, encompassing diverse categories, such as games, utilities, social media, and productivity apps. The dataset comprises textual reviews, ratings, dates, and other relevant metadata.

**Data Preprocessing**: Before diving into the analysis, we perform essential preprocessing steps to clean the textual data. This includes removing special characters, punctuation, stopwords, and applying techniques like tokenization and stemming to normalize the text.

**Exploratory Data Analysis**: We employ various statistical and visual techniques to gain insights into the dataset. This includes analyzing the distribution of ratings, identifying frequently occurring words or phrases, exploring word associations, and investigating trends over time.

**Sentiment Analysis**: To understand the sentiment expressed in the reviews, we utilize sentiment analysis techniques. This involves classifying reviews as positive, negative, or neutral, enabling us to quantify user sentiment and identify patterns across different app categories.

**Correlation Analysis**: We explore potential correlations between user ratings, sentiment, and other relevant variables, such as app category, date, or specific features mentioned in the reviews. This analysis helps identify factors that significantly impact user satisfaction and preferences.

**Visualization**: To enhance our understanding of the data, we create visual representations, such as word clouds, bar plots, heatmaps, and time series plots. These visuals provide an intuitive way to interpret the results and communicate findings effectively.

**Insights and Recommendations**: Based on the analysis outcomes, we derive key insights, such as popular app categories, prevailing sentiment trends, common pain points, and positive aspects highlighted by users. These insights can guide app developers in making informed decisions, improving app design, addressing user concerns, and ultimately enhancing user satisfaction.

As we can see there is only one column named 'Rating' is datatype as float. On observing the dataset we can see that some of columns have integer and float values but the datatype of that column is 'Object'. So to do analysis we have to change the datatype of that column to int/float.

````python
#Coverting 'Price', 'Reviews', and 'Installs' into Numerical values
df1['Price'] = df1['Price'].apply(lambda x: str(x).replace('$', '') if '$' in str(x) else str(x)) 
df1['Price'] = df1['Price'].apply(lambda x: float(x)) 
df1['Reviews'] = df1['Reviews'].apply(lambda x: float(x))
````

Installs: The value of installs is in “string” format. It contains numerical values with commas. It should be removed. And also, the ‘+’ sign should be removed from the end of each string.

````python
#Converting the datatype of 'Installs' column to Float 
df1['Installs'] = df1['Installs'].apply(lambda x: str(x).replace('+', '') if '+' in str(x) else str(x)) 
df1['Installs'] = df1['Installs'].apply(lambda x: str(x).replace(',' , '') if ',' in str(x) else str(x)) 
df1['Installs'] = df1['Installs'].apply(lambda x: float(x))
````

````python
df1['Last Updated']= pd.to_datetime(df1['Last Updated']) 
df1['Last Updated']
````

Size: The size of the app is in “string” format. We need to convert it into a numeric value. If the size is “10M”, then ‘M’ needs to be removed to get the numeric value of ‘10’. If the size is “512k”, which depicts app size in kilobytes, the first ‘k’ should be removed and the size should be converted to an equivalent of ‘megabytes’.

````python
#check unique values of Size column 
df1['Size'].unique()
````

````python
#Checking the type of value present in size column 
type(df1['Size'][0])
````

````python
#As we can see some of values are in 'MB' and some values are in 'k' 
#WE have to convert all the values in 'MB'
#Creating a function which convert all values in 'MB' and get a float value of it.

def convert_to_MB(column):
  if "M" in column: 
    result = float(column.split("M")[0]) 
  elif "k" in column: 
    result = float(column.split("k")[0]) * 0.001 
  elif column == "Varies with device":
    result = np.NaN 
  return result   
````

````python
df1["Size_in_MB"] = df1.Size.apply(convert_to_MB) 
df1["Size_in_MB"].fillna(value=df1.Size_in_MB.mean(), inplace=True)
````

In the graph, reviews with low subjectivity are concentrated at the center of the polarity range [-1, +1] and the reviews with high subjectivity are scattered across the polarity range [-1, +1]. This is understandable because a fact (low subjectivity) is more likely to be neutral (with polarity 0) and an opinion (high subjectivity) is more likely to have a diverse range of negative to positive sentiments.

From the above scatter plot it can be concluded that sentiment subjectivity is not always proportional to sentiment polarity but in maximum number of case, shows a proportional behavior, when variance is too high or low.
